## A. Neutrino Flux from a Ship of Bananas:
Using the decay rate of Potassium-40 and the amount of the bananas and potassium we can estimate the Neutrino Flux.
* Naturally occurring potassium contains about 0.012% of the radioactive isotope ⁴⁰K.
* A typical banana (~118 g) contains about **422 mg of potassium**. Natural potassium includes **0.0117%** of the radioactive isotope **⁴⁰K**, which decays via beta decay (and electron capture), emitting **one neutrino (or antineutrino) per decay**.
- Atoms of ⁴⁰K in a banana:  
  $
  N_{40} \approx 7.6 \times 10^{17}
  $

  $
  \lambda \approx 1.76 \times 10^{-17}\ \text{s}^{-1}
  $
- Activity:

  $
**Result:** A banana produces **≈1.1–1.2 million neutrinos per day**—consistent with the common estimate of **~1 million**.

C. Why Inverse Beta Decay Is Not a Possible Solution?

Inverse Beta Decay (IBD) is a primary and highly effective method for detecting antineutrinos from sources like nuclear reactors, but it is not suitable for detecting those from bananas.​
First let's look through Inverse Beta Decay:
The IBD reaction is: ν̅ₑ + p → e⁺ + n (an electron antineutrino hits a proton, producing a positron and a neutron).​
To create the resulting neutron and positron, the incoming antineutrino must possess a minimum kinetic energy of 1.806 MeV.​
The reason that we can't use the (IBD) to detecting the neutrinos in bananas is that:
The neutrinos from bananas originate from the beta decay of Potassium-40. This decay process releases a total maximum energy (Q-value) of 1.311 MeV, which is shared between the emitted electron and the antineutrino. 

*Consequently, no antineutrino produced by a banana can ever reach the 1.806 MeV threshold required to initiate an inverse beta decay reaction. This energy mismatch makes the detection of banana-based neutrinos via IBD impossible.*

# Exercise 02: Bayesian Parameter Estimation via MCMC

**Context:**  
In observational cosmology, the relationship between redshift $z$ and luminosity distance $D_L$ in a flat $\Lambda$CDM model is approximately:

$$ D_L(z; H_0, \Omega_m) \approx \frac{c(1+z)}{H_0} \int_0^z \frac{dz'}{\sqrt{\Omega_m(1+z')^3 + (1-\Omega_m)}} $$

where $c = 299{,}792.458\ \text{km/s}$.

In this exercise, you will be given synthetic $D_L$ vs. $z$ data generated for realistic cosmological parameters.  
Your task is to use **Metropolis–Hastings MCMC** to estimate $H_0$ and $\Omega_m$ from this noisy dataset.

---

## What to Do

### 1. Model and Likelihood
- Assume Gaussian errors $\sigma_{D_L}$ for each data point.
- The likelihood for parameters $(H_0, \Omega_m)$ is:

$$ L(H_0, \Omega_m) \propto \exp\left( -\frac{1}{2} \sum_{i} \frac{\left[ D_{L,i}^{\text{obs}} - D_L(z_i; H_0, \Omega_m) \right]^2}{\sigma_{D_L}^2} \right) $$

---

### 2. Set Prior Distributions as: 
- $H_0 \sim \mathcal{U}(50,\, 90)$ km/s/Mpc  
- $\Omega_m \sim \mathcal{U}(0.1,\, 0.6)$

---

### 3. Metropolis–Hastings Implementation
- Proposal: Gaussian jumps in both parameters:

$$ H_0^{\text{new}} \sim \mathcal{N}(H_0^{\text{old}}, \delta_{H_0}^2), \quad \Omega_m^{\text{new}} \sim \mathcal{N}(\Omega_m^{\text{old}}, \delta_{\Omega_m}^2) $$

- Compute acceptance ratio:

$$ r = \frac{P_{\text{posterior}}\left(H_0^{\text{new}}, \Omega_m^{\text{new}}\right)}{P_{\text{posterior}}\left(H_0^{\text{old}}, \Omega_m^{\text{old}}\right)} $$

- Accept or reject according to $u < r$, where $u \sim U(0,1)$.

#### Your answer:

**Bayesian Inference:**  
Instead of estimating a single value for a parameter, Bayesian inference yields a **posterior distribution** over possible values, computed via Bayes' theorem:


$ P(\theta \mid D) \propto P(D \mid \theta) \times P(\theta)$ 

- **Posterior** \($P(\theta \mid D)$\): Probability of parameters given the observed data.  
- **Likelihood** \($P(D \mid \theta)$\): Probability of observing the data given the parameters.  
- **Prior** \($P(\theta)$\): Initial belief about the parameters before seeing the data.

**MCMC (Markov Chain Monte Carlo):**  
An algorithm that samples from the posterior distribution by taking random steps through parameter space. This allows us to approximate the shape of the posterior—even when its analytical form is intractable or highly complex.

**Metropolis-Hastings Algorithm:**

1. **Initialize** with a starting parameter value \( $\theta_{\text{old}}$ \).
2. **Propose** a new value \( $\theta_{\text{new}}$ \) using a proposal distribution (e.g., Gaussian).
3. **Compute the acceptance ratio**:
   
   $r = \frac{P(\theta_{\text{new}} \mid D)}{P(\theta_{\text{old}} \mid D)}$
   (Since the posterior is proportional to likelihood × prior, we often compute \( $r = \frac{P(D \mid \theta_{\text{new}}) P(\theta_{\text{new}})}{P(D \mid \theta_{\text{old}}) P(\theta_{\text{old}})}$ \).)
4. **Accept or reject**:  
   - Draw a random number \( $u \sim \text{Uniform}(0, 1)$ \).  
   - If \( $u < r$ \), accept the move: set \( $\theta_{\text{old}} \leftarrow \theta_{\text{new}}$ \).  
   - Otherwise, keep \( $\theta_{\text{old}}$ \) unchanged.
5. **Repeat** thousands of times to generate a Markov chain that samples from the posterior distribution.


```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.integrate import quad
import corner
import seaborn as sns

# Constants
C = 299792.458


# Loading Data
try:
    file_name = "/media/shayan/University/AI-Driven HEP Assignments /assignment-1-Shayan-Zavvarei/src/data/mcmc_cosmo_data.txt"

    data = np.loadtxt(file_name, skiprows=1)
    z_data = data[:, 0]
    d_l = data[:, 1]
    sigma_d = data[:, 2]
    print("data uploaded successfully")
except FileNotFoundError:
    print("Error in uploading data")
    exit()


# Luminosity Distance 
def integrand(z, omega_m):
    return 1.0 / np.sqrt(omega_m * (1+z)**3 + (1-omega_m))

def luminosity_distance(z, h0, omega_m):  
    integral, _ = quad(integrand, 0, z, args=(omega_m,))
    return (C * (1+z) / h0) * integral


# Bayes Functions 
def log_prior(theta):
    h0, Omega_m = theta
    if 50.0 < h0 < 90.0 and 0.10 < Omega_m < 0.60:
        return 0.0
    return -np.inf


def log_likelihood(theta, z, dl, sigma):
    h0, omega_m = theta
    if h0 <= 0 or omega_m < 0:
        return -np.inf 
    dl_model = np.array([luminosity_distance(zi, h0, omega_m) for zi in z])
    chi2 = np.sum(((dl - dl_model) / sigma)**2)
    return -0.5 * chi2


def log_posterior(theta, z, dl, sigma):
    lp = log_prior(theta)
    if not np.isfinite(lp):  
        return -np.inf
    return lp + log_likelihood(theta, z, dl, sigma)


# Metropolis_Hastings MCMC 
def metropolis_hastings(log_posterior_func, n_steps, initial_pos, proposal_width, data_args):
    n_params = len(initial_pos)
    chain = np.zeros((n_steps, n_params))
    chain[0] = initial_pos 
    
    current_log_post = log_posterior_func(initial_pos, *data_args)
    accept_count = 0


    print("program is running")
    for i in range(1, n_steps):
        proposed_pos = chain[i-1] + np.random.normal(0, proposal_width, size=n_params)
        proposed_log_post = log_posterior_func(proposed_pos, *data_args)
        log_acceptance_ratio = proposed_log_post - current_log_post
        if np.log(np.random.rand()) < log_acceptance_ratio:
            chain[i] = proposed_pos
            current_log_post = proposed_log_post
            accept_count += 1  
        else:
            chain[i] = chain[i-1]
    accept_rate = accept_count / (n_steps - 1) 
    print("Finished")
    return chain, accept_rate


# MCMC 
n_steps = 100000
burn_in = 10000
initial_position = np.array([70.0, 0.3])
proposal_widths = np.array([1.5, 0.03])


data_tuple = (z_data, d_l, sigma_d)


chain, acceptance_rate = metropolis_hastings(log_posterior, n_steps, initial_position, proposal_widths, data_tuple)


print(f"Acceptance Rate: {acceptance_rate:.3f}") 
samples = chain[burn_in:]
h0_est, omega_m_est = np.median(samples, axis=0)
h0_std, omega_m_std = np.std(samples, axis=0)


print("\n--- Final Results ---")
print(f"Estimated H0: {h0_est:.2f} ± {h0_std:.2f} km/s/Mpc")
print(f"Estimated Omega_m: {omega_m_est:.3f} ± {omega_m_std:.3f}")


# Plots
print("plotting the corner plot version 1")  
figure = corner.corner(samples, labels=[r"$H_0$", r"$\Omega_m$"],
    quantiles=[0.16, 0.5, 0.84],
    show_titles=True,
    title_kwargs={"fontsize": 12})
figure.suptitle("Posterior Distribution of Cosmological Parameters", fontsize=16)
plt.show()

print("plotting the corner plot version 2")

fig, axes = plt.subplots(2, 2, figsize=(10, 10))


labels = [r"$H_0$", r"$\Omega_m$"]
h0_samples = samples[:, 0]
omega_m_samples = samples[:, 1]


h0_quantiles = np.percentile(h0_samples, [16, 50, 84])
omega_m_quantiles = np.percentile(omega_m_samples, [16, 50, 84])


axes[0, 0].hist(h0_samples, bins=40, density=True, color='steelblue', alpha=0.7, edgecolor='black')
axes[0, 0].axvline(h0_quantiles[1], color='red', linestyle='--', linewidth=2)
axes[0, 0].axvline(h0_quantiles[0], color='orange', linestyle=':', linewidth=1.5)
axes[0, 0].axvline(h0_quantiles[2], color='orange', linestyle=':', linewidth=1.5)
axes[0, 0].set_ylabel('Density', fontsize=12)
axes[0, 0].set_title(f"{labels[0]}\n${h0_quantiles[1]:.2f}^{{+{h0_quantiles[2]-h0_quantiles[1]:.2f}}}_{{-{h0_quantiles[1]-h0_quantiles[0]:.2f}}}$", fontsize=11)
axes[0, 0].tick_params(labelbottom=False)


axes[0, 1].axis('off')


sns.kdeplot(x=h0_samples, y=omega_m_samples, ax=axes[1, 0], 
            fill=True, levels=5, cmap='Blues', alpha=0.6)
axes[1, 0].scatter(h0_samples[::10], omega_m_samples[::10], alpha=0.1, s=1, color='navy')
axes[1, 0].axvline(h0_quantiles[1], color='red', linestyle='--', linewidth=1, alpha=0.5)
axes[1, 0].axhline(omega_m_quantiles[1], color='red', linestyle='--', linewidth=1, alpha=0.5)
axes[1, 0].set_xlabel(labels[0], fontsize=12)
axes[1, 0].set_ylabel(labels[1], fontsize=12)


axes[1, 1].hist(omega_m_samples, bins=40, density=True, orientation='horizontal', 
                color='steelblue', alpha=0.7, edgecolor='black')
axes[1, 1].axhline(omega_m_quantiles[1], color='red', linestyle='--', linewidth=2)
axes[1, 1].axhline(omega_m_quantiles[0], color='orange', linestyle=':', linewidth=1.5)
axes[1, 1].axhline(omega_m_quantiles[2], color='orange', linestyle=':', linewidth=1.5)
axes[1, 1].set_xlabel('Density', fontsize=12)
axes[1, 1].set_title(f"{labels[1]}\n${omega_m_quantiles[1]:.3f}^{{+{omega_m_quantiles[2]-omega_m_quantiles[1]:.3f}}}_{{-{omega_m_quantiles[1]-omega_m_quantiles[0]:.3f}}}$", fontsize=11)
axes[1, 1].tick_params(labelleft=False)

fig.suptitle("Posterior Distribution of Cosmological Parameters", fontsize=16, y=0.995)
plt.tight_layout()
plt.show()

print("Plotting the Trace plot")
fig, axes = plt.subplots(2, figsize=(10, 7), sharex=True)
labels = [r"$H_0$", r"$\Omega_m$"]
for i in range(2):
    ax = axes[i]
    ax.plot(samples[:, i], "k", alpha=0.4)
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)
axes[-1].set_xlabel("Step Number (after Burn-in)")
fig.suptitle("Trace Plots (Chain Convergence Diagnostics)", fontsize=16)
plt.show()
```
